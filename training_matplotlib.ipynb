```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
fig, ax = plt.subplots()
ax.plot([1,2,3,4],[1,4,2,3])
plt.show()
```

```python
x = np.linspace(0, 2, 100)  # Sample data.

# Note that even in the OO-style, we use `.pyplot.figure` to create the Figure.
fig, ax = plt.subplots(figsize=(5, 2.7), layout='constrained')
ax.plot(x, x, label='linear')  # Plot some data on the Axes.
ax.plot(x, x**2, label='quadratic')  # Plot more data on the Axes...
ax.plot(x, x**3, label='cubic')  # ... and some more.
ax.set_xlabel('x label')  # Add an x-label to the Axes.
ax.set_ylabel('y label')  # Add a y-label to the Axes.
ax.set_title("Simple Plot")  # Add a title to the Axes.
ax.legend()  # Add a legend.
```

```python
x = np.linspace(0,10,100)

def f(x):
  return x**(1/2) + 11

plt.figure(figsize=(6,6), layout='constrained')
plt.plot(x, f(x), label='x**(1/2)+11', color='red')
plt.xlabel('x label')
plt.ylabel('y label')
plt.title('testando matplotlib')
plt.legend()
```

```python
plt.figure(figsize=(4,4))
data = np.random.randn(1, 100) # make a random data sets
x = np.linspace(0,2,100) #100 valores entre o e 2
#x and y must be the same size
plt.scatter(x,data, s=50, facecolor='C0', edgecolor='k')
#plt.plot(x,data,'s',label='data')
plt.show()
```

```python
mu, sigma = 115, 15
x = mu + sigma * np.random.randn(10000)
plt.figure(figsize=(5,2.7), layout = 'constrained')
# the histogram of the data
plt.hist(x, 50, density=True, facecolor='C0', alpha=0.75)

plt.xlabel('Length [cm]')
plt.ylabel('Probability')
plt.title('lengths')
plt.text(75, .025, r'$\mu=115,\ \sigma=15$')
plt.axis([55,175,0,0.03])
plt.grid(True)
```

```python
#nao parece muito util
fig, ax = plt.subplots(figsize=(5, 2.7))

t = np.arange(0.0, 5.0, 0.01)
s = np.cos(2 * np.pi * t)
line, = ax.plot(t, s, lw=2)

ax.annotate('local max', xy=(2, 1), xytext=(3, 1.5),
            arrowprops=dict(facecolor='black', shrink=0.05))

ax.set_ylim(-2, 2)
```
